```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.impute import SimpleImputer
```

```python
data = pd.read_csv('drug and alcohol abuse - dataset.csv')
data
```

```python
data.shape
```

```python
data.dtypes
```

```python
data = data.convert_dtypes()
data
```

```python
data.dtypes
```

```python
boolean_columns = data.select_dtypes(include='bool').columns
boolean_columns
```

```python
for col in boolean_columns:
    data[col] = pd.to_numeric(data[col], errors='coerce').astype('Int64')
data
```

```python
data.duplicated()
```

```python
data.duplicated().sum()
```

```python
data.drop_duplicates(inplace=True) # Because we don't have repeated lines, it is not needed to run this line
```

```python
data.drop('Unnamed: 0', axis=1, inplace=True)
data.drop('Phone_Number', axis=1, inplace=True)
```

```python
data.rename(columns={'Does_Sports':'Participates_in_Athletics'}, inplace=True)
```

```python
data.isnull().sum()
```

```python
data["Country"].fillna("Unknown Country", inplace=True)
data["Gender"].fillna("Unknown Gender", inplace=True)
data["School_Type"].fillna("Unknown Status", inplace=True)
```

```python
Age_mean = int(data["Age"].mean())
print(Age_mean)
data["Age"].fillna(Age_mean, inplace= True)

# imputer = SimpleImputer(missing_values=np.nan, strategy='mean')
# column_name = 'Age'
# data[column_name] = imputer.fit_transform(data[[column_name]])
```

```python
imputer = SimpleImputer(missing_values=np.nan, strategy='median')

column_Grade = 'Grades'
data[column_Grade] = imputer.fit_transform(data[[column_Grade]])

column_Family_Income = 'Family_Income'
data[column_Family_Income] = imputer.fit_transform(data[[column_Family_Income]])

# Grades_median = int(data["Grades"].median())
# print(Grades_median)
# data["Grades"].fillna(Grades_median, inplace= True)

# Family_Income_median = int(data["Family_Income"].median())
# print(Grades_median)
# data["Family_Income"].fillna(Family_Income_median, inplace= True)
```

```python
data.isnull().sum()
```

```python
data.dropna(inplace=True)
```

```python
data.isnull().sum()
```

```python
data.reset_index(drop=True, inplace=True)
data
```

## Charts

```python
# pie chart

plt.subplots_adjust(left=0.1,
                    bottom=0.1, 
                    right=0.9, 
                    top=0.9, 
                    wspace=0.4, 
                    hspace=0.4)

plt.subplot(1, 2, 1)
gender_counts.plot.pie(autopct='%1.2f%%', startangle=90, colors=['#66b3ff', '#ff9999'])
plt.title('Distribution of Gender Types')
plt.ylabel('')  # Hide the y-label

plt.subplot(1, 2, 2)
school_type_counts.plot.pie(autopct='%1.2f%%', colors=['#66b3ff', '#ff9999', 'yellow'])
plt.title('Distribution of School Types')
plt.ylabel('')

plt.show()
```

```python
substance_counts = data['Substance_Abuse'].value_counts() # Count the occurrences of each substance abuse type
substance_counts
```

```python
# pie chart
plt.figure(figsize=(10, 6))
myexplode = [0.2, 0, 0, 0] # Explode
substance_counts.plot.pie(autopct='%1.2f%%', colors=['#ff9999','#66b3ff','#99ff99','#ffcc99'], explode = myexplode)
plt.title('Distribution of Substance Abuse Types')
plt.ylabel('')
plt.show()
```

```python
# bar plot
plt.figure(figsize=(10, 6))
sns.countplot(data=data, x='Country', hue='Substance_Abuse', palette='Set2')
plt.title('Country by Substance Abuse')
plt.xlabel('Country')
plt.ylabel('Count')
plt.legend(title='Gender')
plt.grid()
plt.show()
```

```python
# bar plot
plt.figure(figsize=(10, 6))
sns.countplot(data=data, x='Substance_Abuse', hue='Gender', palette='Set2')
plt.title('Substance Abuse by Gender')
plt.xlabel('Substance Abuse Type')
plt.ylabel('Count')
plt.legend(title='Gender')
plt.show()
```

```python
# bar plot
plt.figure(figsize=(10, 6))
sns.countplot(data=data, x='Substance_Abuse', hue='School_Type', palette='Set2')
plt.title('Substance Abuse by Gender')
plt.xlabel('Substance Abuse Type')
plt.ylabel('Count')
plt.legend(title='Gender')
plt.show()
```

```python
# bar plot
plt.figure(figsize=(10, 6))
sns.countplot(data=data, x='Age', hue='Substance_Abuse', palette='Set3')
plt.title('The role of Age in Substance Abuse')
plt.xlabel('Age')
plt.ylabel('Count')
plt.legend(title='Substance Abuse')
plt.grid()
plt.show()
```

```python
# box plot
plt.figure(figsize=(10, 6))
sns.boxplot(data=data, x='Substance_Abuse', y='Grades', hue='Gender', palette='Set2')
plt.title('Grades by Substance Abuse Type')
plt.xlabel('Substance Abuse Type')
plt.ylabel('Grades')
plt.show()
```

```python
# bar plot
data['In_Relationship'] = data['In_Relationship'].astype(str)
plt.figure(figsize=(10, 6))
sns.countplot(data=data, x='Substance_Abuse', hue='In_Relationship', palette='Set2')
plt.title('Substance Abuse by Relationship Status')
plt.xlabel('Substance Abuse Type')
plt.ylabel('Count')
plt.legend(title='In Relationship')
plt.show()
```

```python
# bar plot - countplot
data['Criminal_Record'] = data['Criminal_Record'].astype(str)
plt.figure(figsize=(10, 6))
sns.countplot(data=data, x='Substance_Abuse', hue='Criminal_Record', palette='Set1')
plt.title('Substance Abuse by Criminal Record')
plt.xlabel('Substance Abuse Type')
plt.ylabel('Count')
plt.legend(title='Criminal Record')
plt.show()
```

```python
# box plot
plt.figure(figsize=(10, 6))
sns.boxplot(data=data, x='Substance_Abuse', y='Family_Income', hue='Parents_Together', palette='Set3')
plt.title('Family Income by Substance Abuse Type')
plt.xlabel('Substance Abuse Type')
plt.ylabel('Family Income')
plt.show()
```

```python
# bar plot
data['Parents_Substance_Abuse'] = data['Parents_Substance_Abuse'].astype(str)
plt.figure(figsize=(10, 6))
sns.countplot(data=data, x='Substance_Abuse', hue='Parents_Substance_Abuse', palette='Set2')
plt.title('Substance Abuse by Parents\' Substance Abuse')
plt.xlabel('Substance Abuse Type')
plt.ylabel('Count')
plt.legend(title='Parents Substance Abuse')
plt.show()
```

```python
# distplot
plt.figure(figsize=(10, 6))
sns.distplot(data['Family_Income'], hist=False, kde=True) # kdeplot <= new functions
plt.title('Distribution of Age')
plt.xlabel('Age')
plt.ylabel('Density')
plt.grid(True)
plt.show()
```
