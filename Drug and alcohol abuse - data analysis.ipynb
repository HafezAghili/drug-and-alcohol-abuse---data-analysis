```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.impute import SimpleImputer
```

```python
data = pd.read_csv('drug and alcohol abuse - dataset.csv')
data
```

```python
data.shape
```

```python
data.dtypes
```

```python
data = data.convert_dtypes()
data
```

```python
data.dtypes
```

```python
boolean_columns = data.select_dtypes(include='bool').columns
boolean_columns
```

```python
for col in boolean_columns:
    data[col] = pd.to_numeric(data[col], errors='coerce').astype('Int64')
data
```

```python
data.duplicated()
```

```python
data.duplicated().sum()
```

```python
data.drop_duplicates(inplace=True) # Because we don't have repeated lines, it is not needed to run this line
```

```python
data.drop('Unnamed: 0', axis=1, inplace=True)
data.drop('Phone_Number', axis=1, inplace=True)
```

```python
data.rename(columns={'Does_Sports':'Participates_in_Athletics'}, inplace=True)
```

```python
data.isnull().sum()
```

```python
data["Country"].fillna("Unknown Country", inplace=True)
data["Gender"].fillna("Unknown Gender", inplace=True)
data["School_Type"].fillna("Unknown Status", inplace=True)
```

```python
Age_mean = int(data["Age"].mean())
print(Age_mean)
data["Age"].fillna(Age_mean, inplace= True)

# imputer = SimpleImputer(missing_values=np.nan, strategy='mean')
# column_name = 'Age'
# data[column_name] = imputer.fit_transform(data[[column_name]])
```

```python
imputer = SimpleImputer(missing_values=np.nan, strategy='median')

column_Grade = 'Grades'
data[column_Grade] = imputer.fit_transform(data[[column_Grade]])

column_Family_Income = 'Family_Income'
data[column_Family_Income] = imputer.fit_transform(data[[column_Family_Income]])

# Grades_median = int(data["Grades"].median())
# print(Grades_median)
# data["Grades"].fillna(Grades_median, inplace= True)

# Family_Income_median = int(data["Family_Income"].median())
# print(Grades_median)
# data["Family_Income"].fillna(Family_Income_median, inplace= True)
```

```python
data.isnull().sum()
```

```python
data.dropna(inplace=True)
```

```python
data.isnull().sum()
```

```python
data.reset_index(drop=True, inplace=True)
data
```

## Charts

```python
# pie chart

plt.subplots_adjust(left=0.1,
                    bottom=0.1, 
                    right=0.9, 
                    top=0.9, 
                    wspace=0.4, 
                    hspace=0.4)

plt.subplot(1, 2, 1)
gender_counts.plot.pie(autopct='%1.2f%%', startangle=90, colors=['#66b3ff', '#ff9999'])
plt.title('Distribution of Gender Types')
plt.ylabel('')  # Hide the y-label

plt.subplot(1, 2, 2)
school_type_counts.plot.pie(autopct='%1.2f%%', colors=['#66b3ff', '#ff9999', 'yellow'])
plt.title('Distribution of School Types')
plt.ylabel('')

plt.show()
```

```python
substance_counts = data['Substance_Abuse'].value_counts() # Count the occurrences of each substance abuse type
substance_counts
```

```python
# for corr() we need to have numeric data -> we map non-numerical data to numerical data
data["Gender_code"] = np.where(data["Gender"] == "Female", 0, 1)
data.head()
```

```python
data_for_cor = data.drop(['Country', 'Gender', 'Substance_Abuse', 'School_Type'], axis=1)
data_for_cor
```

```python
cor=data_for_cor.corr() # for corr() we need to have numeric data -> we remove other non-numerical data , correlation -> It shows the correlation between the data
cor
```

```python
sns.heatmap(cor, xticklabels=cor.columns, yticklabels=cor.columns, vmin=-1, vmax=1)
```

```python
from scipy.stats import pearsonr # We use it where the data is continuous
```

```python
p_coeff, p_value = pearsonr(data.Grades, data.Family_Income) # The effect of family income and students' grades on each other
```

```python
p_coeff
```

```python
from scipy.stats import spearmanr # We use it where the data is discrete
```

```python
p_coeff, p_value = spearmanr(data.In_Relationship, data.Participates_in_Athletics) # The effect of being an athlete and being in a relationship on each other
p_coeff
```
